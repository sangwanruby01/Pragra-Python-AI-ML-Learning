(1️) User Authentication System
Create:
Class: User
● username
● password (private)
● role (Admin / Manager / Employee)
Methods:
● login()
● change_password()
● validate_user()
Store users in a JSON file.

```python
                                           ## 1) User Authentication System ##

import json         #Import JSON

class User:
    def __init__(self, username,password,role):
        accepted_roles=["admin","employee","manager"]          #Allowed roles

        if role not in accepted_roles:                            #Role validation
            raise ValueError(f"Accepted roles are only: {accepted_roles}")

        self.username = username
        self.__password = password            # Role validation
        self.role = role
        self.attempts=0                      # wrong login attempts counter

    def login(self, username, password, max_attempt=3):              #Login method with attempt limit
        if self.attempts >= max_attempt:
            return "Account locked! Too many attempts!"
        if username == self.username and password == self.__password:
            self.attempts = 0                                       #reset attempts if login successfully
            return "Login Successfully!"
        else:
            self.attempts += 1
            return f"Incorrect username or password! Attempt {self.attempts}/{max_attempt}"

    def change_password(self, old_password, new_password):           #Change password method
        if self.__password == old_password:
            self.__password = new_password
            return "password changed successfully!"
        else:
            return "Incorrect old password!"

    def validate_user(self,username,password,role):                   #Validate user method on the basis od username,password,role
        if (self.username == username and
            self.__password == password and
            self.role == role):
            return True
        else:
            return False

try:
    with open("users.json", "r") as file:           #Load users from JSON file
        users = json.load(file)
except:
    users = []                                     #if file missing or corrupted, creating new list
    print("File not found or corrupted")

with open("users.json", "w") as file:               #store users in JSON file
    json.dump(users, file, indent=4)
```

(2) Employee Management Module
Class: Employee
Attributes:
● employee_id
● name
● department
● salary
● joining_date
Methods:
● update_salary()
● calculate_annual_salary()
● display_details()

```python
                                           ## 2) Employee Management Module ##

import csv             #Import csv

class Employee:
    def __init__(self, employee_id, name, department, salary, date_of_joining):
        self.employee_id = employee_id
        self.name = name
        self.department = department
        self.salary = salary
        self.date_of_joining = date_of_joining

    def update_salary(self, new_salary):          #Update salary method
        if new_salary < 0:
            return "Salary cannot be negative"
        self.salary = new_salary
        return "Salary updated successfully"

    def calculate_annual_salary(self):           #Calculate yearly salary method
        return self.salary * 12

    def display_details(self):                   #Display details in dictionary format
        return {
            "Employee ID": self.employee_id,
            "Name": self.name,
            "Department": self.department,
            "Salary": self.salary,
            "Joining Date": self.date_of_joining
        }

emp1 = Employee(1999610, "Riya", "Floor Ready", 7000, "02-03-2021")
emp2 = Employee(2108461, "Mark", "Bulk", 6000, "05-07-2020")

with open("employees.csv", "w", newline="") as file:                         #Store employees records in CSV file
    writer = csv.writer(file)

    writer.writerow(["Employee ID", "Name", "Department", "Salary", "Joining Date"])     #column names

    writer.writerow([emp1.employee_id, emp1.name, emp1.department, emp1.salary, emp1.date_of_joining])
    writer.writerow([emp2.employee_id, emp2.name, emp2.department, emp2.salary, emp2.date_of_joining])

print("Data saved in employees.csv")
```

(3) Department Module
Class: Department
Attributes:
● dept_id
● dept_name
● employee_list (list of employee IDs)
Methods:
● add_employee()
● remove_employee()
● total_salary_expense()

```python
                                           ## 3) Department Module ##

class Department:
    def __init__(self, dept_id, dept_name):
        self.dept_id = dept_id
        self.dept_name = dept_name
        self.employee_list = []

    def add_employee(self, employee_id):             #Add Employee method, handle duplicates
        if employee_id in self.employee_list:
            return "Employee already in department"
        self.employee_list.append(employee_id)
        return "Employee added"

    def remove_employee(self, employee_id):        #Remove employee method, handle non-existing
        if employee_id in self.employee_list:
            self.employee_list.remove(employee_id)
            return "Employee removed"
        return "Employee not found"

    def total_salary_expense(self, employees):     # Calculate total salary of employees
        total = 0
        for emp in employees:
            if emp.employee_id in self.employee_list:
                total += emp.salary
        return total
```

(4️) Payroll System
Class: Payroll
Methods:
● calculate_tax()
● generate_salary_slip()
● deduct_leave()
Should:
● Apply tax slabs
● Deduct based on leave days
● Generate salary breakdown

```python
                                           ## 4) Payroll System ##

class Payroll:
    def __init__(self, employee, leave_days):
        self.employee = employee
        self.leave_days = leave_days
        self.daily_salary = self.employee.salary / 30         # daily salary calculation
        self.leave_deduction = self.deduct_leave()            # leave deduction
        self.tax = self.calculate_tax()                       # tax calculation
        self.net_salary = self.employee.salary - self.leave_deduction - self.tax    #net salary calculation

    def calculate_tax(self):                     #Calculate tax based on salary
        salary = self.employee.salary

        if salary <= 5000:
            return salary * 0.02                  #tax 2%
        elif salary <= 8000:
            return salary * 0.05                 #tax 5%
        else:
            return salary * 0.08                 #tax 8%

    def deduct_leave(self):                      ## leave deduction calculation
        if self.leave_days < 0:
            return 0
        else:
            return self.leave_days * self.daily_salary   # leave deduction

    def salary_breakdown(self):             #Salary breakdown details
        return {
            "Employee ID": self.employee.employee_id,
            "Name": self.employee.name,
            "Department": self.employee.department,
            "Gross Salary": self.employee.salary,
            "Leave Deduction": self.leave_deduction,
            "Tax Deducted": self.tax,
            "Net Salary": self.net_salary
        }
```

(5) Project Assignment Module
Class: Project
Attributes:
● project_id
● project_name
● assigned_employees (list)
Methods:
● assign_employee()
● remove_employee()
● list_project_members()


```python
                                          ## 5) Project Assignment Module ##
class Project:
    def __init__(self, project_id, project_name):
        self.project_id = project_id
        self.project_name = project_name
        self.assigned_employees = []        #empty list of assigned employees

    def assign_employee(self, employee_id):             #assign employee to project
        if employee_id in self.assigned_employees:                     # checking duplicate
            return "Employee already assigned to this project"

        self.assigned_employees.append(employee_id)
        return "Employee assigned"

    def remove_employee(self, employee_id):                   # remove employee from project
        if employee_id in self.assigned_employees:
            self.assigned_employees.remove(employee_id)
            return "Employee removed"
        return "Employee not found in project"

    def list_project_members(self):                  # list all assigned employees
        return self.assigned_employees
```

(6️) Reporting System
Generate reports:
● Total employees
● Highest paid employee
● Department-wise salary expense
● Active projects
● Monthly payroll summary
Use dictionaries + loops heavily here.


```python
                                         ## 6) Reporting System ##

class Reports:
    def __init__(self, employees, department, project):
        self.employees = employees
        self.department = department
        self.project = project

    def total_employees(self):        #Total number of employees
        return len(self.employees)

    def highest_paid_employee(self):          #Highest Paid employee
        if not self.employees:
            return "No employees"

        highest = self.employees[0]
        for emp in self.employees:
            if emp.salary > highest.salary:
                highest = emp

        return {
            "Employee ID": highest.employee_id,
            "Name": highest.name,
            "Salary": highest.salary
        }

    def department_salary_expense(self):            #Department-wise salary expense
        expense = {}

        for d in self.department:
            total = 0
            for emp in self.employees:
                if emp.employee_id in d.employee_list:
                    total += emp.salary
            expense[d.dept_name] = total

        return expense

    def active_projects(self):                 #Active projects
        active = {}

        for p in self.project:
            active[p.project_name] = p.list_project_members()

        return active

    def monthly_payroll_summary(self):                #Monthly payroll summary
        summary = {}

        for emp in self.employees:
            summary[emp.employee_id] = {
                "Name": emp.name,
                "Monthly Salary": emp.salary,
                "Tax": emp.salary * 0.05,
                "Net Salary": emp.salary - (emp.salary * 0.05)
            }

        return summary
```
