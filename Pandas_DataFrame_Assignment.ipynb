(16️) Find customers whose PremiumAmount is above the average for their PolicyType.

```python
avg_premium = insurance.groupby("PolicyType")["PremiumAmount"].mean()

insurance["AvgPremium"] = insurance["PolicyType"].map(avg_premium)

above_avg = insurance[insurance["PremiumAmount"] > insurance["AvgPremium"]]
```

(17️) Find the top 3 customers with the highest ClaimAmount.

```python
high_claim_amount=insurance.sort_values(by="ClaimAmount",ascending=False,inplace=False)
high_claim_amount.head(3)
```

(18️) Create a new column RiskLevel:
● “High” if ClaimAmount > 2000 ● “Medium” if between 500–2000 ● “Low” otherwise.

```python
insurance["RiskLevel"] = "Low"

insurance.loc[(insurance["ClaimAmount"] >= 500) & (insurance["ClaimAmount"] <= 2000), "RiskLevel"] = "Medium"

insurance.loc[insurance["ClaimAmount"] > 2000, "RiskLevel"] = "High"
```

(19️) Filter customers where: ○ TenureYears > 5 and ○ NoOfClaims >= 1 and ○ PolicyType = 'Life'

```python
insurance[(insurance["TenureYears"]>5) & (insurance["NoOfClaims"]>=1) & (insurance["PolicyType"]=="Life")]
```

```python
sum_of_premium_amount_by_region=insurance.groupby("Region")["PremiumAmount"].sum()

region_with_highest_premium_amount = sum_of_premium_amount_by_region.idxmax()
```
