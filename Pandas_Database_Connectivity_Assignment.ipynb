Q2️  Read Data into Pandas
Read the entire employees table into a Pandas DataFrame.
● Display the first 5 rows and DataFrame info (.info() and .describe()).
● Confirm correct data types for each column.

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Pragra123@localhost:3306/company_db")

employees_df = pd.read_sql("SELECT * FROM employees", con=engine)          #read the entire employees table

print(employees_df.head())    #display first 5 rows
```

```python
employees_df.describe()         #display summary statistics
```

Q3️ — Filtering Data
From the DataFrame:
● Display employees who have experience > 5 years and salary > 60000.
● Display only name, department, and salary columns for these filtered employees.

```python
emp = employees_df[(employees_df["experience"] > 5) & (employees_df["salary"] > 60000)]
emp[["name","department","salary"]]          #display only name, department and salary columns
```

Q4️ — Aggregation and Grouping
Use Pandas groupby() and agg() to find:
● Average salary by department
● Maximum experience per location
● Total salary expenditure per department

```python
employees_df.groupby("department")["salary"].mean()    #average salary by department
```

```python
employees_df.groupby("location")["experience"].max()    #max experience per location
```

```python
employees_df.groupby("department")["salary"].sum()       #Salary expenditure per department
```

Q5️ — Data Modification and New Columns
Add a new derived column named performance_level:
● “High” if salary > 80000
● “Medium” if salary between 50000 and 80000
● “Low” otherwise
Display only name, department, salary, and performance_level.

```python
employees_df["performance_level"]="Low"

employees_df.loc[employees_df["salary"] > 80000, "performance_level"] = "High"

employees_df.loc[
    (employees_df["salary"] >= 50000) & (employees_df["salary"] <= 80000), "performance_level"] = "Medium"

employees_df[["name", "department", "salary", "performance_level"]]    #display only name, department, salary and performance_level columns
```

Q6️ - Export Results
Save your final processed DataFrame (after all transformations) to:
● A new MySQL table named employee_summary, and
● A CSV file named employee_summary.csv. (Use to_sql() and to_csv() methods.)

```python
employees_df.to_sql(name="employee_summary", con=engine, if_exists="replace", index=False)   #save dataframe to a new MySQL table
```

```python
employees_df.to_csv("employee_summary.csv", index=False)        #save dataframe to csv file
```
